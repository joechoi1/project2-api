## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
#save and print the text content of a page with all tags removed

import requests
from bs4 import BeautifulSoup


url = 'https://en.wikipedia.org/wiki/Bear'
response = requests.get(url)


if response.status_code == 200:
    # Parse the page with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Get the text content and print it
    # Use strip=True to remove leading and trailing spaces
    page_text = soup.get_text(strip=True, separator=' ')  # Adding a space as a separator for readability
    print(page_text)
else:
    print("Failed to fetch the page, status code:", response.status_code)


```

```python
#download an image with beautifulsoup and save it in this repository
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin


url = 'https://en.wikipedia.org/wiki/Bear'
response = requests.get(url)


if response.status_code == 200:
   
    soup = BeautifulSoup(response.text, 'html.parser')

    
    bear_image_element = soup.find('img', alt=lambda alt: alt and 'bear' in alt.lower())

    if bear_image_element:
       
        image_url = bear_image_element['src']

        
        full_image_url = urljoin(url, image_url)

        print(f"Bear Image URL: {full_image_url}")
        
       
        image_response = requests.get(full_image_url)
        
        if image_response.status_code == 200:
            
            image_path = 'bear_image.jpg'
            with open(image_path, 'wb') as f:
                f.write(image_response.content)
            print(f"Bear image downloaded and saved as '{image_path}'")
        else:
            print(f"Failed to download the image, status code: {image_response.status_code}")
    else:
        print("No bear image found on the page.")
else:
    print(f"Failed to fetch the page, status code: {response.status_code}")
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
import requests
from bs4 import BeautifulSoup


url = 'https://en.wikipedia.org/wiki/Bear'
response = requests.get(url)


if response.status_code == 200:
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
  
    all_links = soup.find_all('a', href=True)

    
    for i, link in enumerate(all_links[:10]):  
        href = link['href']
        print(f"Link {i+1}: {href[:100]}")
else:
    print(f"Failed to fetch the page, status code: {response.status_code}")
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
import pandas as pd
import requests
from bs4 import BeautifulSoup


scripts = pd.read_csv('pudding_data.csv')
links = scripts['link']  


for i, link in enumerate(links):
    try:
        
        response = requests.get(link)
        
        
        if response.status_code == 200:
            
            soup = BeautifulSoup(response.text, 'html.parser')
            
            
            script_text = soup.get_text(separator=" ", strip=True)
            
            
            first_1000_chars = script_text[:1000]
            
          
            print(f"Script {i+1} from {link} (first 1000 characters):")
            print(first_1000_chars)
            print("-" * 80)  
        else:
            print(f"Failed to fetch the script from {link}, status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred with {link}: {e}")

```

```python
#download the movie posters for 10 of these movies and save them to this repository
import os

poster_folder = 'posters'
os.makedirs(poster_folder, exist_ok=True)

for movie in data['results']:
    poster_path = movie.get('poster_path')
    poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"

    poster_name = movie.get('title').replace(" ", "_") + '_poster.jpg'
    poster_path_local = os.path.join(poster_folder, poster_name)

    response = requests.get(poster_url)

    with open(poster_path_local, "wb") as file:
        file.write(response.content)
```
